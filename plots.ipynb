```python
import numpy as np

# Define the original array
x = np.linspace(0, 1, 100)  # 100 points between 0 and 1

# Calculate the original step sizes
steps_original = np.diff(x)

# Randomly perturb each step size
noise_factor = 0.005  # Adjust this factor to control the amount of randomization
randomized_steps = steps_original * (1 + noise_factor * np.random.randn(len(steps_original)))

# Calculate the new points based on the randomized step sizes
randomized_x = np.cumsum(randomized_steps)
randomized_x = np.concatenate([[0], randomized_x])  # Add the starting point back

# Verify the unique step values and the number of elements
print("Unique step values (randomized):", np.unique(randomized_steps))
print("Number of elements:", len(randomized_x))

x=randomized_x
x
```

# Four-way Dist Plots

```python
# smoothen and plot:

def smooth_data(MotionData, col_name, normalize=False, interpolate=True):
    """
    Takes the file dataframe and desired column to smooth (i.e. thumb01x)
    iterates through each element of the column (angle data) and smooths any
    jumps in the data with a constant being the difference between before and
    after the jump. Comment out the second last line to disable normalizing.
    """
    data = MotionData[col_name].copy()  # Create a copy of the column data
    i = 1
    while i < len(data):
      if abs(data[i-1] - data[i]) > 10:  # Jump detected
        constant = data[i-1] - data[i]
        data[i:] += constant
      i += 1

    if normalize: data -= data[0]
    MotionData[col_name] = data  # Update the column in the original DataFrame

def smooth_ratio(MotionData, col_name, normalize=False, interpolate=True):
    """
    Takes the file dataframe and desired column to smooth (i.e. thumb01x)
    iterates through each element of the column (angle data) and smooths any
    jumps in the data with a constant being the difference between before and
    after the jump. Comment out the second last line to disable normalizing.
    """
    data = MotionData[col_name].copy()  # Create a copy of the column data
    i = 1
    while i < len(data):
      if abs(data[i-1] - data[i]) > 0.005:  # Jump detected
        constant = data[i-1] - data[i]
        data[i:] += constant
      i += 1

    if normalize: data -= data[0]
    MotionData[col_name] = data  # Update the column in the original DataFrame

def smooth_file(file_name, cols = None, normalize=False):
    """
    repeats smoothening for all desired columns (3 axes of 2 measured fingers)
    """
    #print(f"reading {data_path+file_name}")
    MotionData = pd.read_csv(data_path+file_name)
    if cols is None:
      for finger in ['Thumb', 'Index']:
        for axis in ['x', 'y', 'z']:
          smooth_data(MotionData, finger+'01'+axis, normalize)
    else:
      for col in cols:
          smooth_data(MotionData, col, normalize)
    return MotionData

# general func to compute dist(base, joint2) / dist(base, joint1)
def compute_ratio(file, base, joint1, joint2, output):
  if isinstance(file, str):
    df = pd.read_csv(data_path+file)
  else: df = file
  shape = df['Handx'].shape
  for joint in [base, joint1, joint2]:
    for axis in ['x', 'y', 'z']:
      smooth_data(df, joint+axis)

  base_dfs = [df[base+'x'],df[base+'y'],df[base+'z']]
  joint1_dfs = [df[joint1+'x'],df[joint1+'y'],df[joint1+'z']]
  joint2_dfs = [df[joint2+'x'],df[joint2+'y'],df[joint2+'z']]

  dist1 = np.sqrt((joint1_dfs[0] - base_dfs[0])**2 + (joint1_dfs[1] - base_dfs[1])**2 + (joint1_dfs[2] - base_dfs[2])**2)
  dist2 = np.sqrt((joint2_dfs[0] - base_dfs[0])**2 + (joint2_dfs[1] - base_dfs[1])**2 + (joint2_dfs[2] - base_dfs[2])**2)

  base = base[0:-1] if base[-1] == '_' else base
  joint1 = joint1[0:-1] if joint1[-1] == '_' else joint1
  joint2 = joint2[0:-1] if joint2[-1] == '_' else joint2

  df[base + "" + joint1 + "_dist"] = dist1
  smooth_ratio(df, base + "" + joint1 + "_dist")
  df[base + "" + joint2 + "_dist"] = dist2
  smooth_ratio(df, base + "" + joint2 + "_dist")
  df[output] = np.where(dist1 != 0, dist2 / dist1, 0)
  smooth_ratio(df, output)
  return df, (base + "" + joint1 + "_dist"), (base + "" + joint2 + "_dist")
```

```python
cols = ['Thumb01x', 'Thumb01y', 'Thumb01z', 'Index01x', 'Index01y', 'Index01z', 'Hand_posThumbCMC_dist', 'Hand_posThumbMCP_dist', 'Hand_posIndexMCP_dist', 'Hand_posIndexPIP_dist', 'ThumbMCPIndexMCP_dist', 'ThumbMCPIndexPIP_dist', 'Thumb_Hand_dist_ratio', 'Index_Hand_dist_ratio', 'Index_Thumb_dist_ratio']
force_cols = [' hook', ' paper', ' pencilThumb',' pencilFinger',' bottleVert', ' bottleRadial',' sphereThumb', ' sphereFL']
cols = cols + force_cols

def process_file(path):
  df = pd.read_csv(data_path+path)

  for col_name in ['Thumb01', 'Index01','Hand_pos_', 'ThumbCMC', 'ThumbMCP','IndexMCP', 'IndexPIP']:
    for axis in ['x', 'y', 'z']:
      smooth_data(df, col_name+axis, normalize=False)

  for col_name in force_cols:
    smooth_data(df, col_name, normalize=False)

  # compute ratios
  df, _,_ = compute_ratio(df, 'Hand_pos_', 'ThumbCMC', 'ThumbMCP', 'Thumb_Hand_dist_ratio')
  df, _,_ = compute_ratio(df, 'Hand_pos_', 'IndexMCP', 'IndexPIP', 'Index_Hand_dist_ratio')
  df, _,_ = compute_ratio(df, 'ThumbMCP', 'IndexMCP', 'IndexPIP', 'Index_Thumb_dist_ratio')

  #print('after compute_ratio: ' + str(list(df.columns)))

  # 1. clip the data
  df, picked_col= trim_by_biggest_range(df)

  # set timestamp min-max to 0 to 2 with sklearn
  df['Timestamp'] = minmax_scale(df[['Timestamp']])

  # 2. interpolate, 0.01 = 100 cycles per second, at 2 sec
  df.loc[:, 'Timestamp'] -= df['Timestamp'].iloc[0] # normalize time
  interp_prefft_duration = 2
  interp_prefft_step = 0.01
  interp_data = pd.DataFrame()
  interp_prefft_time = np.linspace(0, interp_prefft_duration, num=int(interp_prefft_duration/interp_prefft_step)+1)
  interp_data['Timestamp'] = interp_prefft_time
  for col_name in cols:
      interp_data[col_name] = np.interp(interp_prefft_time, df['Timestamp'], df[col_name])

  return interp_data, picked_col

def trim_by_biggest_range(df):
  #print("trimming!")
  # trim by second derivative
  second_dev_upper_threshold = 25000
  second_dev_lower_threshold = -1 * second_dev_upper_threshold

  min_trim_array_length = 10  # minimum no. of datapoints after trimming

  def find_trim_indices(data, above_threshold, below_threshold):
      data_array = np.array(data)
      start_idx = np.argmax(data_array > above_threshold) + 1
      end_idx = len(data_array) - np.argmax(data_array[::-1] < below_threshold) + 1
      #print(f"trim: {start_idx}-{end_idx} instead of 0-{len(df['Timestamp'])}")

      # Ensure start_idx is less than end_idx
      if start_idx >= end_idx:
          return 0, len(data_array)

      return start_idx, end_idx

  columns_to_check = ['Thumb01x', 'Thumb01y', 'Thumb01z', 'Index01x', 'Index01y', 'Index01z']
  ranges = {col: df[col].max() - df[col].min() for col in columns_to_check}
  max_range_column = max(ranges, key=ranges.get)
  #print(f"picking {max_range_column}")

  gradient = np.gradient(df[max_range_column], df['Timestamp'])
  second_dev = np.gradient(gradient, df['Timestamp'])

  # Find trim indices based on second_dev thresholds
  start_idx, end_idx = find_trim_indices(second_dev, second_dev_upper_threshold, second_dev_lower_threshold)

  # Trim data
  trimmed_df = df.iloc[start_idx:end_idx+1]
  for c in cols:
    if 'ratio' in c:
      trimmed_df.loc[:, c] -= trimmed_df[c].iloc[0] - 1 #normalize to 1
    else:
      trimmed_df.loc[:, c] -= trimmed_df[c].iloc[0]  # normalize to 0
    # leave dist alone
  #print(list(trimmed_df.columns))
  #trimmed_df['Timestamp'] = minmax_scale(trimmed_df[['Timestamp']])
  #print('after trimming: ' + str(list(trimmed_df.columns)))
  return trimmed_df, max_range_column
```

```python
col = 'Index01z'
path = 'FMA17Massfingerflexion_LeftHand_07_09_2024_12_06_27.csv'
data_path = '/content/drive/My Drive/Bionics Lab Research/Grace/highschool-data/JS_Score2/'


for threshold in [0.25, 0.275, 0.3, 0.4, 0.5]:
  df = pd.read_csv(data_path+path)

  # smoothen cols
  for col_name in ['Thumb01', 'Index01','Hand_pos_', 'ThumbCMC', 'ThumbMCP','IndexMCP', 'IndexPIP']:
    for axis in ['x', 'y', 'z']:
      smooth_data(df, col_name+axis, normalize=False)
  for col_name in force_cols:
    smooth_data(df, col_name, normalize=False)

  # compute ratios
  df, _,_ = compute_ratio(df, 'Hand_pos_', 'ThumbCMC', 'ThumbMCP', 'Thumb_Hand_dist_ratio')
  df, _,_ = compute_ratio(df, 'Hand_pos_', 'IndexMCP', 'IndexPIP', 'Index_Hand_dist_ratio')
  df, _,_ = compute_ratio(df, 'ThumbMCP', 'IndexMCP', 'IndexPIP', 'Index_Thumb_dist_ratio')

  # normalize cols
  for c in cols:
    if '01' in c: df.loc[:, c] -= df[c].iloc[0] #normalize to 0
    else: df.loc[:, c] -= df[c].iloc[0] + 1  # normalize to 1

  og_data = df.copy()

  # clip the data
  df, picked_col, second_dev, second_dev_upper_threshold, second_dev_lower_threshold = trim_by_biggest_range(df, threshold, col)
  clip_data = df.copy()

  # set timestamp min-max to 0 to 2 with sklearn
  df['Timestamp'] = minmax_scale(df[['Timestamp']], feature_range=(0, 2))
  minmax_data = df.copy()

  # interpolate, 0.01 = 100 cycles per second, at 2 sec
  df.loc[:, 'Timestamp'] -= df['Timestamp'].iloc[0] # normalize time
  interp_prefft_duration = 2
  interp_prefft_step = 0.01
  interp_data = pd.DataFrame()
  interp_prefft_time = np.linspace(0, interp_prefft_duration, num=int(interp_prefft_duration/interp_prefft_step)+1)
  interp_data['Timestamp'] = interp_prefft_time
  for col_name in cols:
      interp_data[col_name] = np.interp(interp_prefft_time, df['Timestamp'], df[col_name])

  # Choose scale_factor as the smaller value to ensure both conditions are satisfied
  scale_factor = min(50 / max(second_dev), -50 / min(second_dev))
  second_dev_upper_threshold *= scale_factor
  second_dev_lower_threshold *= scale_factor
  second_dev *= scale_factor

  plt.figure(figsize=(8, 5))  # Optional: set the figure size
  plt.plot(og_data['Timestamp'], og_data[col], marker='o', linestyle='-', color='r', label='original')
  plt.plot(clip_data['Timestamp'], clip_data[col], marker='o', linestyle='-', color='y', label='clipped data')
  plt.plot(minmax_data['Timestamp'], minmax_data[col], linestyle='-', color='g', label='minmax data to 0-2')
  plt.plot(og_data['Timestamp'], second_dev, marker='o', linestyle='-', color='y', label='second_dev', alpha=0.2)
  plt.fill_between(og_data['Timestamp'], second_dev, where=(second_dev > second_dev_upper_threshold), interpolate=True, color='g')
  plt.fill_between(og_data['Timestamp'], second_dev, where=(second_dev < second_dev_lower_threshold), interpolate=True, color='g')
  plt.title(col+', clipped at threshold='+str(threshold))
  plt.xlabel('Timestamp')
  plt.ylabel(col)
  plt.grid(True)
  plt.legend()
  plt.show()
```
